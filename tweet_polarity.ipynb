```python
import csv
import nltk

from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')
```

```python
sid = SentimentIntensityAnalyzer()
with \
        open('DownsampledCoreDataWithoutBots.csv',
             'r+',
             encoding="utf-8") as inp, \
        open('DownsampledCoreDataPolarityWithoutBots.csv',
             'w+',
             encoding="utf-8") as out:

    my_fields = ['text',
                 'migration_relevance',
                 'compound_polarity',
                 'negative_polarity',
                 'neutral_polarity',
                 'positive_polarity']

    writer = csv.DictWriter(out, fieldnames=my_fields)
    writer.writeheader()

    reader = csv.DictReader(inp)

    for row in reader:
        print(row['text'])
        ss = sid.polarity_scores(row['text'])
        writer.writerow({'text': str(row['text']),
                         'migration_relevance': row['migration_relevance'],
                         'compound_polarity': ss['compound'],
                         'negative_polarity': ss['neg'],
                         'neutral_polarity': ss['neu'],
                         'positive_polarity': ss['pos']})

        for k in sorted(ss):
            print('{0}: {1}, '.format(k, ss[k]), end='')
        print()
```
