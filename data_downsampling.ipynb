```python
import pandas as pd
from sklearn.utils import resample

df = pd.read_csv('CoreData.csv')
df['migration_relevance'] = df['migration_relevant'].map({'yes': 1, 'no': 0, "Yes": 1, "No": 0})
vcount = df['migration_relevance'].value_counts()

print(vcount)

df_majority = df[df.migration_relevance == 0.0]
df_minority = df[df.migration_relevance == 1.0]
df_majority_downsampled = resample(df_majority,
                                   replace=False,
                                   n_samples=300,
                                   random_state=123)
df_downsampled = pd.concat([df_majority_downsampled, df_minority])

print(df_downsampled.migration_relevance.value_counts())
print(df_downsampled)

df_downsampled.to_csv("DownsampledCoreData.csv", index=False)
```
